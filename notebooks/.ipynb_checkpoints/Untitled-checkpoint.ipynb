```python
def extract_window(activity_map: np.ndarray, x: int, y: int, support: float):
    new_map = deepcopy(activity_map)
    support = support // 2
    
    # add columns (or rows) if on the edge
    if x < support:
        x = support
        new_map = np.concatenate((np.repeat(new_map[:, 0][:, np.newaxis], support - x, axis=1), new_map), axis=1)

    elif x > new_map.shape[0] - support:
        x = new_map.shape[0] - support
        new_map = np.concatenate(
            (new_map, np.repeat(new_map[:, -1][:, np.newaxis], new_map.shape[0] - support + x, axis=1)), axis=1)

    if y < support:
        y = support
        new_map = np.concatenate((np.repeat(new_map[0, :][np.newaxis, :], support - y, axis=0), new_map), axis=0)

    elif y > new_map.shape[1] - support:
        y = new_map.shape[1] - support
        new_map = np.concatenate(
            (new_map, np.repeat(new_map[-1, :][np.newaxis, :], new_map.shape[0] - support + y, axis=0)), axis=0)

    window = new_map[x-support: x+support+1, y-support: y+support+1]
    return window
```

```python
def boundary_detection_activity(x, y):
    minus_supp = 9 // 2
    plus_supp = 5 // 2
    new_v = deepcopy(model.preprocessing_stage[1].V)
    
    window = extract_window(new_v, x, y, 9)
    smooth = ndimage.gaussian_filter(window, 2)
    raw_activity = np.abs(window - smooth)
    return ndimage.gaussian_filter(extract_window(raw_activity, minus_supp, minus_supp, 5), 1)[2,2]

#     return ndimage.gaussian_filter(raw_activity[minus_supp-plus_supp:minus_supp+plus_supp+1, minus_supp-plus_supp:minus_supp+plus_supp+1], 1)[2,2]
#     return raw_activity[minus_supp-plus_supp:minus_supp+plus_supp+1, minus_supp-plus_supp:minus_supp+plus_supp+1][2,2]
# plt.imshow(boundary_detection_activity(0,0)) 
# 
# new_v = model.preprocessing_stage[1].V
# np.concatenate((np.array(new_v[:0,]*4), new_v), axis=1)
# column: new_v[:,0][:, np.newaxis]
# row: new_v[0,:][np.newaxis, :]
# ax[0].imshow(new_v)

vb = np.vectorize(boundary_detection_activity)
space = np.arange(0,121)
Y, X = np.meshgrid(space, space)
res = vb(X,Y)
# res = []
# for x in range(0,121):
#     tmp = []
#     for y in range(0,121):
#         tmp.append(boundary_detection_activity(x,y))
#     res.append(tmp)
# res = np.array(res)
fig, axs = plt.subplots(1,2, figsize=(20,10))
axs[0].imshow(na)
axs[1].imshow(res)
# boundary_detection_activity(50,60)
```

```python
def gauss2D_iso_cart(x, y, mu=(0.0, 0.0), sigma=1.0, normalize_RFs=False):
    """gauss2D_iso_cart
    gauss2D_iso_cart takes two-dimensional arrays x and y, containing
    the x and y coordinates at which to evaluate the 2D isotropic gaussian 
    function, with a given sigma, and returns a 2D array of Z values.
    Parameters
    ----------
    x : numpy.ndarray, 2D or flattened by masking
        2D, containing x coordinates
    y : numpy.ndarray, 2D or flattened by masking
        2D, containing y coordinates
    mu : tuple, optional
        mean, 2D coordinates of mean/mode of gauss (the default is (0.0,0.0))
    sigma : float, optional
        standard deviation of gauss (the default is 1.0)
    Returns 
    -------
    numpy.ndarray, 2D or flattened by masking
        gaussian values evaluated at (x,y)
    """
    if normalize_RFs:
        return np.exp(-((x-mu[0])**2 + (y-mu[1])**2)/(2*sigma**2)) /(2*np.pi*sigma**2)
    else:
        return np.exp(-((x-mu[0])**2 + (y-mu[1])**2)/(2*sigma**2))
```
